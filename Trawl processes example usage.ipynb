## 1. Simulation

Consider a homogeneous LÃ©vy basis $L$, a (potentially unbounded) set of finite area $A$ and a collection of time-indexed sets $A_t$ given by
\begin{equation*}
    A_t = A + (t,0), \qquad A = \{(s,x) \in \mathbb{R}^2 \colon  s<0, 0 < x < \phi(s) \},
\end{equation*}
where $\phi \colon (-\infty,0] \to \mathbb{R}_{\ge 0}$ is an increasing, convex and continuous function. We refer to $\phi$ as the trawl function and to $A$ as the trawl set. We aim to simulate the trawl process $$X_t= L(A_t) = L(A + (t,0))$$ at given times $\tau_1, \ldots,\tau_k$ by simulating the Gaussian and jump parts separately. For the slice partition, we assume that the times are equidistant, i.e. $\tau_1 =\tau, \tau_2 = 2 \tau, \ldots, \tau_k = k \tau$. If the times are not equidistant, please consider 
* employing the grid or cpp algorithms
* modifying the  `compute_slice_areas_finite_decorrelation_time` / `compute_slice_areas_infinite_decorrelation_time` methods

We first instantiate a trawl class and then use the `simulate` class method.

`trawl_instance.simulate(method)`

where the input `method` is one of the strings `slice,grid` or `cpp` (note that the word method here is first used as a class method then as an input to the class method). All the three simulation methods requires the following arguments:`nr_simulations` and `trawl_function`.
        
        
### 1.1 Slice method

In the slice method, we consider equidistant times $\tau_1,\ldots,\tau_k = \tau, \ldots,k \tau.$ If the ambit set $A$ is compactly supported, we set `decorrelation_time` $= \sup\left\{T <0: \phi(T) =0\right\},$ otherwise we set it be $-\infty.$

In the following, we let $\tau=0.2,k=1000, \phi \colon [-1,0] \to \mathbb{R}, \phi(x) = (1+x)$ and `decorrelation_time`$=-1$. 
Then $\mathrm{Leb}(A) = 0.5$ and the marginal distribution of this trawl process is given by the sum of independent random variables distributed as
$$\mathcal{N}(-3\ \textrm{Leb}(A),7 \ \textrm{Leb}(A)^{0.5}) + \text{Gamma}(2\  \textrm{Leb}(A),3) = \mathcal{N}(-1.5,3.5 \sqrt{2}) + Gamma(1,3) $$

where the distributions are parametrised as in [scipy](https://docs.scipy.org/doc/scipy/reference/stats.html). We perform $50$ simulations. 

```python
tau = 0.2
nr_trawls = 1000  #this is k
nr_simulations = 50
trawl_function_slice_1 = lambda x :  (x>-1) * (x<=0) * (1+x) 
decorrelation_time_slice_1 = -1
gaussian_part_params = (-3,7)
jump_part_params = (2,3)
jump_part_name   = 'gamma' #can be None,'cauchy','invgauss','poisson' (more to be implemented)
```

The required parameters are `nr_trawls, nr_simulations, trawl_function, tau, decorrelation_time, gaussian_part_params,
jump_part_name` and  `jump_part_params`.

```python
trawl_slice_1 = trawl(nr_trawls = nr_trawls, nr_simulations = nr_simulations,trawl_function = trawl_function_slice_1,tau =  tau,
                   decorrelation_time =  decorrelation_time_slice_1, gaussian_part_params = gaussian_part_params,
                   jump_part_name =  jump_part_name,jump_part_params = jump_part_params)   
trawl_slice_1.simulate(method='slice')
```

The  simulated values are stored in the `values`,`gaussian_values` and `jump_values` attributes of the trawl class, which have shapes given by [`nr_simulations`,`nr_tralws`]. The `values` attribute is the sum of the other two attributes.

```python
np.array_equal(trawl_slice_1.values, trawl_slice_1.gaussian_values + trawl_slice_1.jump_values)
```

```python
f,ax = plt.subplots(1,2,figsize=(14,6))
ax[0].plot(trawl_slice_1.values[0])
ax[1].plot(trawl_slice_1.values[1])
f.suptitle('Realisations of the trawl process')
```

### 1.2  Grid method

In this algorithm, we discretize the rectangle $[\tau_1+\text{truncation_grid},\tau_k] \times [0,\phi(0)]$ into a grid of cells, simulate the LÃ©vy basis over these cells and approximate $L(A)$ by $\sum{L(c)},$ where the sum is over cells $c$ which are contained in $A.$  If the ambit set $A$ is bounded, the parameter `truncation_grid` is the same as `decorrelation_time` from the slice algorithm, otherwise `truncation_grid`  dictates how much of the ambit set we neglect (see figure from ). The finer the `mesh_size`, the smaller the approximation error of the simulation.

```python
nr_simulations = 50
trawl_function_grid = lambda x :  (x> -2) * (x<=0) * (2 - (-x) **2/2) 
gaussian_part_params = (-3,7)
jump_part_params = (2,3)
jump_part_name   = 'gamma'
mesh_size = 0.05
truncation_grid = -2
times_grid =  tau * np.arange(1,nr_trawls+1,1)#doesn't have to be equally spaced, but have to be strictly increasing
```

```python
trawl_grid  = trawl(nr_simulations = nr_simulations,trawl_function = trawl_function_grid,
                   times_grid=times_grid,mesh_size = mesh_size,truncation_grid = truncation_grid,
                   gaussian_part_params = gaussian_part_params, jump_part_name =  jump_part_name,
                   jump_part_params = jump_part_params )
```

```python
trawl_grid.simulate(method='grid')
```

```python
f,ax = plt.subplots(1,2,figsize=(14,6))
ax[0].plot(trawl_grid.values[0])
ax[1].plot(trawl_grid.values[1])
f.suptitle('Realisations of the trawl process')

#Even if we have the same parameters for the Gaussian and jump part, the marginal distribution of the trawl process is 
#different, because the Lebesgue measure of the ambit set is different. 
```

We can check that the marginal distribution of the simulated trawls agrees with the theoretical ones. We infer the distribution of the Gaussian and jump parts via mle, then adjust the obatined value by dividing by the area of the ambit set. We plot the histogram obtained below.

```python
a = [norm.fit(data = trawl_grid.gaussian_values[simulation,:]) for simulation in range(trawl_grid.nr_simulations)]
total_area_grid = quad(trawl_grid.trawl_function,a=-np.inf,b=0)[0]
a = np.array(a) / np.array([total_area_grid, total_area_grid ** 0.5])
    
f,ax= plt.subplots(1,2,sharey=True,figsize=(14,6))
ax[0].hist(a[:,0],density=True)
ax[0].set_title('infered means and true value')
    
ax[1].hist(a[:,1],density=True)
ax[1].set_title('infered sd and true value')
    
ax[0].axvline(x=trawl_grid.gaussian_part_params[0],color='b')
ax[1].axvline(x=trawl_grid.gaussian_part_params[1],color='b')
f.suptitle('Gaussian params')
#true vlaue is shown in blue
```

```python
a = [gamma.fit(data = simulation,floc=0) for simulation in trawl_grid.jump_values]
a = np.array([[i[0],i[2]] for i in a])   #a, scale
a = a / np.array([total_area_grid,1])
        
f_gamma,ax_gamma= plt.subplots(1,2,sharey=True, figsize=(14,6))
ax_gamma[0].hist(a[:,0],density=True)
ax_gamma[0].set_title('infered shape and true value')
        
ax_gamma[1].hist(a[:,1],density=True)
ax_gamma[1].set_title('infered scale and true value')
        
ax_gamma[0].axvline(x=trawl_grid.jump_part_params[0],color='b')
ax_gamma[1].axvline(x=trawl_grid.jump_part_params[1],color='b')
f_gamma.suptitle('Gamma parameters')
#true vlaue is shown in blue
```

### 1.3 Cpp simulation

The parameters required for the cpp simulation algorithm are `nr_simulations, trawl_function, cpp_truncation, cpp_times, cpp_part_name, cpp_part_params` and `cpp_intensity`.

The text below is from section $3.2$ of [Integer Valued Trawls](https://scholar.harvard.edu/files/GeneralTrawlPaper.pdf).  

Suppose that the LÃ©vy seed satisfies $L^{\prime} \sim N B(m, \theta)$, i.e. the corresponding probability mass function is given by
$$
P\left\{L^{\prime}=j\right\}=\frac{1}{j !} \frac{\Gamma(m+j)}{\Gamma(m)}(1-\theta)^{m} \theta^{j}, \quad \text { for } j \in \mathbb{N}_{0}
$$
and let the LÃ©vy process $\left(L_{t}^{\prime}\right)_{t \geq 0}$ associated with the LÃ©vy seed through $L^{\prime} \stackrel{\text { law }}{=} L_{1}^{\prime}.$ Then $\left(L_{t}^{\prime}\right)_{t \geq 0}$ satisfies $L_{t}^{\prime} \sim N B(m t, \theta)$. We recall that the negative binomial LÃ©vy process may be thought of as a compound Poisson process $L_{t}^{\prime}=\sum_{i=1}^{N_{t}} C_{i}$ where $N$ is a Poisson process with intensity $v=$ $m|\log (1-\theta)|$ and the innovations $C_{i}$ follow the logarithmic distribution with point probabilities
$$
\mathbb{P}\left(C_{i}=j\right)=\frac{1}{|\log (1-\theta)|} \frac{\theta^{j}}{j}, \quad \text { for } j \in \mathbb{N} .
$$
The above representation result will prove useful in the context of simulating a negative Binomial trawl process, see Section 7 of [Integer Valued Trawls](https://scholar.harvard.edu/files/GeneralTrawlPaper.pdf). 
Recall that the cumulant function of the negative Binomial LÃ©vy seed is given by
$$
C\left(\zeta, L^{\prime}\right)=m\left(\log (1-\theta)-\log \left(1-\theta e^{i \zeta}\right)\right)
$$
which implies that
$$
X_{t}=L\left(A_{t}\right) \sim N B(\operatorname{leb}(A) m, \theta) .
$$

In the following, we specify the distribution of $C_i$ through `cpp_part_name` and `cpp_part_params`, the intensity measure through `cpp_intensity`, the times at which we simulate the trawl processes through `cpp_times` and the truncation parameter through `cpp_truncation`. In particular, we disregard parts of the ambit sets contained in the part of the plane $t<\text{min(cpp_times) + cpp_truncation} $.

```python
trawl_function =  lambda x : np.exp(x) * (x<=0)
cpp_truncation = -30
cpp_part_name = 'logser' # see https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.logser.html
cpp_part_params = (0.6,)
m=10
#not required for the simulation. we only use it to have the same number of samples in the qqplot from the next cell
nr_trawls_cpp  = 2500 
cpp_intensity =  m * abs(np.log(1 - cpp_part_params[0]))
cpp_times     = 0.15 * np.arange(1,nr_trawls_cpp+1,1)


trawl_cpp = trawl(nr_simulations   = nr_simulations, trawl_function=trawl_function,
                cpp_truncation  = cpp_truncation, cpp_times = cpp_times, cpp_part_name = cpp_part_name,
                cpp_part_params = cpp_part_params,cpp_intensity = cpp_intensity)
                
trawl_cpp.simulate(method='cpp')
```

As seen from the qqplots and the histogram, we conclude that the empirical distribution of the trawl process agrees with the theoretical one.

```python
from statsmodels.graphics.gofplots import qqplot_2samples
from scipy.stats import nbinom
total_area_cpp = quad(trawl_cpp.trawl_function,a=-np.inf,b=0)[0]

val1 = trawl_cpp.values[5]
val2 = nbinom.rvs(m*total_area_cpp ,  1-cpp_part_params[0],size=nr_trawls_cpp)
qqplot_2samples(val1,val2,line='45')

#if the spacing between the ambit sets is small, the samples might be very corrrelated and 
#the qqplot might not be representative. so it's somewhat expected for the y=x line to not exactly match the qqplot
#when the nr of trawls is small
```

```python
bins = list(range(math.ceil(max(trawl_cpp.values[2]))))
rv = nbinom(m * total_area_cpp,1-cpp_part_params[0])
yy = [rv.pmf(i) for i in bins]

f_bins,ax_bins = plt.subplots()
ax_bins.hist(trawl_cpp.values[2], density=True, bins = bins, alpha = 0.5)
ax_bins.scatter(bins,yy,c='k',s=10)
_ = ax_bins.set_title('Empirical histogram of simulated trawl values (red) against theoretical pmf (black)')
```

#### Scaling
Gaussian LÃ©vy basis

Let $L^{'}\sim \mathcal{N}(\mu,\sigma)$ be Gaussian distributed with mean $\mu$ and variance $\sigma^2.$ Then $X_t = L(A_t) \sim \mathcal{N}\left(\mu \mathrm{Leb}\left(A\right), \sigma^2 \mathrm{Leb}\left(A\right)\right).$


Gamma LÃ©vy basis
Let $L^{'} \sim \text{Gamma}(\alpha,\beta)$ with shape and rate parameters $\alpha,\beta >0$ and pdf
\begin{equation*}
    p(x) = \frac{\beta^\alpha}{\Gamma(\alpha)} x^{\alpha-1} e^{-\beta x}, \ x>0.
\end{equation*}
Then $X_t = L(A_t) \sim \text{Gamma}(\alpha \mathrm{Leb}\left(A\right),\beta).$
